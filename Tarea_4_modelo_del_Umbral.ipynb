$\textbf{Modelo del Umbral}$

El "Modelo del Umbral" o "Modelo de Granovetter" busca caracterizar y simular las dinamica de una manisfestacion multitudinal, En este modelo, cada individuo tiene un umbral de adopción, que es un nivel mínimo de influencia necesaria para que tomen una decisión específica, como cambiar de opinión, adoptar una nueva conducta o unirse a un grupo.

los individuos tienen de forma particular cada uno un Umbral definido de cuantos otros individuos es necesario que sumen a una tendencia para que este tambien se una a esta tendencia , se estudian 2 formas de distribuir el umbral dentro de una población de N personas.

La simulación consistita  en una función que dispone la cantidad de personas que se suman a la protesta, cuyo comportamiento cambia dependiendo de si los umbrales de los agentes del sistema están distribuidos de forma normal o uniforme.

```python
#MODELO DEL UMBRAL-----------------------------------

import numpy as np
import matplotlib.pyplot as plt
 
def modelo_umbral(distribucion,N=1000,umbrales=100): #distribucion corresponde al tipo de dist que se quiere estudiar
    umb = np.linspace(0,100,umbrales)                #N es el numero total de individuos
    manifestantes = []                                          
    for i in range(umbrales):                           
      amarillos = 0
      movilizados = 0
      for j in range(N):
        if distribucion[j]>i:#se chequea la condicion para unirse a la manifestacion en funcion de cada individuo
          amarillos += 1
        else:
          movilizados += 1
      manifestantes.append(movilizados/10)

    plt.figure(figsize=(10,7))
    plt.plot(umb,umb)
    plt.plot(umb, manifestantes, color='orange')
    plt.title('Modelo del umbral distribucion normal')
    plt.xlabel('umbral')
    plt.ylabel('adhesiones')
    plt.grid()
    
    

dis_normal = np.abs(np.random.normal(0.4,0.2,N))*100
modelo_umbral(dis_normal)
```
