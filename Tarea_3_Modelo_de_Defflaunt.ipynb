$\textbf{Modelo de Deffaunt}$

el modelo de Deffaunt que funcionamente similarmente al modelo del votante pero en este caso la opinion  individuo se encuentra en un continuo definido entre 0 y 1 segun el modelo. Definicion que mantiene la binariedad en el sentido de que 0 y 1 en este contexto siguen siendo opiniones totalmente opuestas, en funcion de esta hipotesis una población de $N$ individuos $i$ con una opinión en un espectro continuo $x_i$, en cada step dos individuos comparten y de cierta manera ''promedian'' su opinión cuando la diferencia entre sus opiniones es menor que un umbral $d$ de la siguiente forma:  

$$
\begin{array}{lr} x & = x +\mu(x'-x)\\ x' & = x' +\mu(x-x') \end{array}
$$

en donde se cumple que  $|x-x^{´}| \leq d $.

$\textbf{Simulacion}$: La simulación describe el modelo de Deffaunt en una sola dimensión, que luego es evaluada y graficada, bajo las variables definidas anteriormente, cosiderando la inforacion reportadad en el articulo de referncia. 

```python
#SIMULACION DEFFAUNT
import numpy as np
import matplotlib.pyplot as plt 
from datetime import datetime
from tqdm.auto import tqdm

#Condiciones iniciales 
d = 0.5
mu = 0.5
N=100
n_pasos=5000
agentes = list(np.random.uniform(0, 1, size=N))#Se define un espectro de opinion que va de 0 a 1 y subdividido en 1000
opiniones_totales=np.zeros((n_pasos,N))
for n in range(n_pasos):
      i, j = np.random.randint(N), np.random.randint(N)
      if np.abs(agentes[i]- agentes[j])<d:#Verifica que la diferencia de opiniones de los indv cumpla con la distancia escogida
        #Si los individuos interactuan, se calcula el cambio segun el paper
        agentes[i] = agentes[i] + mu*(agentes[j] - agentes[i])
        agentes[j] = agentes[j] + mu*(agentes[i] - agentes[j])
        opiniones_totales[n,:] = agentes.copy()
      else:#Si la diferencia de opinion es mayor a la diferencia escogida los individuos no interactuan
        pass

tiempos = [[i]*N for i in range(n_pasos)]




plt.figure(figsize=(10,7))
plt.scatter(tiempos,opiniones_totales, s=5, color='red')
plt.xlabel("Tiempo")
plt.ylabel("Opinión")
plt.title("Estado de consenso")
plt.grid()
plt.show()

```
